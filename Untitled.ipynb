## Mengubah null value

```python
# isi AlasanResign dengan masih_bekerja karena belum resign
df1['AlasanResign'] = df1['AlasanResign'].fillna('masih_bekerja')
# menggunakan nilai 0 pada JumlahKeikutsertaanProjek, JumlahKeterlambatanSebulanTerakhir, dan JumlahKetidakhadiran
df1['JumlahKetidakhadiran'] = df1['JumlahKetidakhadiran'].fillna(df1['JumlahKetidakhadiran'].median())
df1['JumlahKeikutsertaanProjek'] = df1['JumlahKeikutsertaanProjek'].fillna(df1['JumlahKeikutsertaanProjek'].median())
df1['JumlahKeterlambatanSebulanTerakhir'] = df1['JumlahKeterlambatanSebulanTerakhir'].fillna(df1['JumlahKeterlambatanSebulanTerakhir'].median())
# menggunakan nilai mode untuk SkorKepuasanPegawai
df1['SkorKepuasanPegawai'] = df1['SkorKepuasanPegawai'].fillna(df1['SkorKepuasanPegawai'].mode()[0])
```

## Mengubah dtypes yg tidak sesuai

```python
df1['SkorKepuasanPegawai'] = df1['SkorKepuasanPegawai'].astype(int)
df1['JumlahKeikutsertaanProjek'] = df1['JumlahKeikutsertaanProjek'].astype(int)
df1['JumlahKeterlambatanSebulanTerakhir'] = df1['JumlahKeterlambatanSebulanTerakhir'].astype(int)
df1['JumlahKetidakhadiran'] = df1['JumlahKetidakhadiran'].astype(int)
```

```python
df1['TanggalLahir'] = pd.to_datetime(df1['TanggalLahir'])
df1['TanggalHiring'] = pd.to_datetime(df1['TanggalHiring'])
df1['TanggalPenilaianKaryawan'] = pd.to_datetime(df1['TanggalPenilaianKaryawan'])
df1['TanggalResign'] = pd.to_datetime(df1['TanggalResign'], errors = 'coerce')
```
